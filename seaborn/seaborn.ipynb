```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('tips') 
data
```

```python
data.head()
```

```python
data['total_bill']
```

```python
sns.lineplot(data['total_bill'])
```

```python
sns.scatterplot(x='total_bill',y='tip',data = data)
```

```python
sns.histplot(data['total_bill'])
```

```python
sns.kdeplot(data['total_bill'])
```

```python
#catplot
#relplot
```

```python
sns.relplot(x="total_bill",y="tip",data=data)
```

```python
data['time'].unique()
```

```python
sns.relplot(x="total_bill",y="tip",hue='time',style="smoker",data=data)
```

```python
sns.relplot(x="total_bill",y="tip",hue='time',style="smoker",kind="scatter",row="time",data=data)
```

```python
sns.relplot(y="total_bill",x="tip",hue="time",style='smoker',kind="scatter",col="smoker",data=data)
```

```python
data['day'].unique()
```

```python
sns.relplot(y = "total_bill", x = "tip", hue = 'time', data = data ,kind = "scatter", col = 'day')  # this is a scatter plot and y is the total bill and x is the tip and data is the data set and hue is the time and col is the day 
```

```python
sns.relplot(y = 'total_bill', x = 'tip', data = data )
plt.title('This is a scatter plot')
plt.xlabel('Total bill', fontsize = 20, color= 'black')
plt.ylabel('Tip', fontsize = 20, color= 'green')
plt.show()
```
