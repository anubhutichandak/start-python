```python
x = np.array([1,2,3,4])
mylabels = ['apples','bananas','cherries','dates']
plt.pie(x, labels = mylabels, startangle=90)  
plt.show()
```

```python
x = np.array([1,2,3,4])
mylabels = ['apples','bananas','cherries','dates']
myexplode = [0.2,0,0,0]
plt.pie(x, labels = mylabels, startangle=90, explode=myexplode)  
plt.show()
```

```python
x = np.array([1,2,3,4])
mylabels = ['apples','bananas','cherries','dates']
myexplode = [0.2,0,0,0]
plt.pie(x, labels = mylabels, startangle=90, explode=myexplode, shadow=True)  
plt.show()
```

```python
x = np.array([1,2,3,4])
mylabels = ['apples','bananas','cherries','dates']
colour = ['red','green','blue','yellow']
myexplode = [0.2,0,0,0]
plt.pie(x, labels = mylabels, startangle=90, explode=myexplode, shadow=True, colors=colour)  
plt.show()
```

```python
x = np.array([1,2,3,4])
mylabels = ['apples','bananas','cherries','dates']
colour = ['red','green','blue','yellow']
myexplode = [0.2,0,0,0]
plt.pie(x, labels = mylabels, startangle=90, explode=myexplode, shadow=True, colors=colour, autopct='%1.1f%%')  
plt.show()
```

```python
x = np.array([1,2,3,4])
mylabels = ['apples','bananas','cherries','dates']
colour = ['red','green','blue','yellow']
myexplode = [0.2,0,0,0]
plt.pie(x, labels = mylabels, startangle=90, explode=myexplode, shadow=True, colors=colour, autopct='%1.1f%%')  
plt.legend(title = 'My pie chart')
plt.show()
```
